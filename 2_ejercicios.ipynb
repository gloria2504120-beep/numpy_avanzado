```python
import numpy as np
```

```python
#Ejercicio 1
import numpy as np
# 1. Definimos el arreglo
a = np.array([1, 2, 3, 4])
resultado = np.all(a != 0)

print(f"¿Es verdad que ninguno de los elementos es cero?: {resultado}")# ejercicio 1
```

```python
#Ejercicio 2
hay_ceros = np.any(a == 0)

resultado_alterno = not hay_ceros
print(f"Resultado con lógica inversa: {resultado_alterno}")
```

```python
#Ejercicio 3

a = np.array([1, 2, np.nan, np.inf])

contiene_nan = np.isnan(a).any()

contiene_inf = np.isinf(a).any()

print(f"¿Contiene valores NaN?: {contiene_nan}")
print(f"¿Contiene valores infinitos?: {contiene_inf}")

print(f"Máscara de NaNs: {np.isnan(a)}")
print(f"Máscara de Infs: {np.isinf(a)}")
```

```python
#ejercicio 4
a = np.array([1, 2, np.inf])

resultado = np.isinf(a).any()

print(f"¿El arreglo contiene algún valor infinito?: {resultado}")

posicion = np.where(np.isinf(a))
print(f"El infinito se encuentra en el índice: {posicion[0]}")
```

```python
#Ejercicio 5
a = np.array([1, np.nan, 3])
resultado = np.isnan(a).any()

print(f"¿El arreglo contiene valores NaN?: {resultado}")
cuantos_nan = np.isnan(a).sum()
print(f"Cantidad de valores faltantes (NaN): {cuantos_nan}")
```

```python
#Ejercicio 6
# 1. Definimos los arreglos
a = np.array([1, 3, 5])
b = np.array([2, 3, 4])

mayor = a > b
mayor_o_igual = a >= b
menor = a < b
menor_o_igual = a <= b

print(f"¿a > b?  : {mayor}")         
print(f"¿a >= b? : {mayor_o_igual}") 
print(f"¿a < b?  : {menor}")         
print(f"¿a <= b? : {menor_o_igual}") 
```

```python
#Ejercicio 7
a = np.array([1, 7, 13, 105])
memoria_total = a.nbytes
print(f"Arreglo: {a}")
print(f"Tipo de dato: {a.dtype}")
print(f"Cantidad de elementos: {a.size}")
print(f"Memoria total ocupada: {memoria_total} bytes")
```

```python
#Ejercicio 8
ceros = np.zeros(10)
unos = np.ones(10)
cincos = np.ones(10) * 5

print(f"Ceros:  {ceros}")
print(f"Unos:   {unos}")
print(f"Cincos: {cincos}")
```

```python
#Ejercicio 19

m = np.array([[1, 2], [3, 4]])
m_booleanos = (m % 2 == 0)

m_solo_pares = m[m % 2 == 0]

print("Matriz original:")
print(m)
print("\nMatriz de verificación (Booleanos):")
print(m_booleanos)
print("\nValores que cumplieron la condición:")
print(m_solo_pares)
```
